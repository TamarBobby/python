```python
import numpy as np
```

```python
x = np.linspace(0, 10, 1000) # 100 נקודות בין 0 ל-10
y = np.sin(4*x +6)
```

```python
plt.plot(x, y, label=' y = 4x + 6')
plt.title('Basic Line Graph')
plt.grid()  
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

```python
x = np.linspace(0, 10, 100) # 100 נקודות בין 0 ל-10
y = np.sin((10*x)**2+2*x)
```

```python
plt.plot(x, y, label=' y = 10X^2 + 2X')
plt.title('Basic Line Graph')
plt.grid()  
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

```python
years = range(2010, 2021)
consumption_oil = [100, 105, 110, 120, 125, 130, 135, 140, 145, 150, 155]
consumption_gas = [80, 82, 85, 88, 90, 95, 100, 105, 110, 115, 120]
consumption_renewable = [20, 25, 30, 35, 40, 45, 50, 60, 70, 80, 90]


fig, axs = plt.subplots(2, 2)  # יוצר מטריצה של גרפים בגודל 2x2
axs[0, 0].plot(years, consumption_oil, 'tab:blue')
axs[0, 0].set_title('oil')
axs[0, 1].plot(years, consumption_gas, 'tab:red')
axs[0, 1].set_title('gas')
axs[1, 0].scatter(years, consumption_renewable, c='yellow')
axs[1, 0].set_title('renewable')
axs[1, 1].scatter([1,2,3],[1,2,3], c='pink')
axs[1, 1].set_title('TAMAR')

# התאמת הגרפים למראה נקי יותר
for ax in axs.flat:
    ax.label_outer()
```
